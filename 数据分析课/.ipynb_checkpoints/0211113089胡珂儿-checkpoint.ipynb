```python
import pandas as pd
import numpy as np
data = pd.read_csv('/Users/hukeer/Desktop/上海楼房.csv')
data
```

```python
data.info()
```

### 去除价格未能正常呈现的数据

```python
data = data[data['价格'].str.match('^[0-9]+$')]
data
```

### 去除不是上海地区的

```python
quyu =  ['黄浦','杨浦','虹口','普陀','静安','浦东','闵行','松江','宝山','青浦','嘉定','奉贤','崇明','金山']
data = data[data['位置'].isin(quyu)]
data
```

### 求各区房价均值

```python
result = data.groupby(by='位置',as_index=False)['价格'].mean()
result
```

### 求出房价均价最高的是哪个区，且房价是多少

```python
result.sort_values(by='价格',ascending=False,inplace=True)
result
```

### 求出每个区有多少个楼盘

```python
data.groupby(by='位置')['楼盘名称'].count()
```
