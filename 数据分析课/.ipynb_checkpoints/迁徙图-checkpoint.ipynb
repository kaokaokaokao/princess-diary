```python
from pyecharts.charts import Geo
import pyecharts.options as opts
from pyecharts.globals import ChartType, SymbolType
from pyecharts.commons.utils import JsCode
```

```python
x = ['广州', '上海', '新疆', '黑龙江', '北京']
y = [10000, 7000, 1500, 1000, 500]
lst = [(i,j)for i,j in zip(x,y)]
start_end = [('广州','上海'),('广州','新疆'),('广州','黑龙江'),('广州','北京')]
```

```python
geo = (
    Geo()
    .add_schema('china', itemstyle_opts=opts.ItemStyleOpts(color='#222222', border_color='red'))
    .add('', lst, type_=ChartType.EFFECT_SCATTER,
         label_opts=opts.LabelOpts(formatter=JsCode(
                """function(params){
                if ('value' in params.data){
                return params.data.name + ':' + params.data.value[2];
                }
            }"""            )
)
        )
    .add('', start_end, 
         type_=ChartType.LINES,
         effect_opts=opts.EffectOpts(symbol=SymbolType.ARROW,color='yellow', symbol_size=8)
        )
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(min_=500, max_=11000))
)
geo.render_notebook()
```
