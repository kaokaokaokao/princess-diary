```python
turnover = pd.read_csv('/Users/hukeer/Desktop/函数分析.csv')
turnover
```

```python
print(turnover.std())
print(turnover.cov())
```

```python
turnover['总销售额'] = turnover.loc[:,'化妆品':].sum(axis=1)
turnover
```

```python
turnover['平均销售额'] = turnover.loc[:,'化妆品':'食品'].mean(axis=1)
turnover
```

```python
turnover[turnover['总销售额']<turnover['总销售额'].quantile(q=0.3)]
```

```python
turnover.loc[:,['化妆品']].mean()
```
