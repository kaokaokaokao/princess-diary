```python
import jieba
txt = ''
for i in comments:
    txt += i

lst = jieba.lcut(txt)

lst
```

```python
with open('stopwords.txt') as f:
    stop_words = eval(f.read())
    
nlst = []
for word in lst:
    if word not in stop_words:
        nlst.append(word)

s = pd.Series(nlst)
ns = s.value_counts()
ns
```

```python
import matplotlib.pyplot as plt
import wordcloud
img = plt.imread('ditu.jpg')
w = wordcloud.WordCloud(font_path='msyh.ttc', mask=img, max_words=1300,margin=3,width=1800,height=800,random_state=42)  
w.generate_from_frequencies(ns)
w.to_file('aa.jpg')
```
