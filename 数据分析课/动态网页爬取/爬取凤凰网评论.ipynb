```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
driver.get('https://gentie.ifeng.com/c/comment/84sMdkBWv5s')
#注意这里一定要安装谷歌驱动插件，在此网址http://chromedriver.storage.googleapis.com/index.html下载符合自己谷歌浏览器的版本。
#如果是其他浏览器打开，也可用对应浏览器插件，如driver = webdriver.Edge()
```

```python
from lxml import etree
def get_data_x(html):
    dom = etree.HTML(html)
    lst_x = dom.xpath('//p[@class="content_txt-3sJKgW6Y"]/text()')
    return lst_x
```

```python
import time
import  random
lst = []
wait = WebDriverWait(driver, 10)
while True:
    if driver.find_element_by_css_selector( '#root > div.mainContent-3HwfGFG1 > div.commentPage-lpELXu_5 > div.commentLeft-6AsK2zsC > div.comment_box-1dciHGm9 > div:nth-child(3) > div.hotList-3nzzINWZ > div:nth-child(20) > p.content_txt-3sJKgW6Y')!=[]:
        html = driver.page_source
        data = get_data_x(html)
        lst.extend(data)
    else:
        break
    if driver.find_element_by_css_selector('#root > div.mainContent-3HwfGFG1 > div.commentPage-lpELXu_5 > div.commentLeft-6AsK2zsC > div.comment_box-1dciHGm9 > div.paginationBox-3JBiOrPn > a:nth-child(12)')==[]:
        break

    wait.until(
        EC.element_to_be_clickable(
            (By.CSS_SELECTOR, '#root > div.mainContent-3HwfGFG1 > div.commentPage-lpELXu_5 > div.commentLeft-6AsK2zsC > div.comment_box-1dciHGm9 > div.paginationBox-3JBiOrPn > a:nth-child(12)')
        )
    ).click()
    time.sleep(random.randint(3,8))
```
