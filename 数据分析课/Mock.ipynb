```python
from pyecharts.charts import Scatter,Bar,Line
import pyecharts.options as opts
```

```python
x = data['favorite']
y = data['coin']
scatter = (
    Scatter(init_opts=opts.InitOpts(width="720px", height="320px"))
    .add_xaxis(xaxis_data=data['favorite'])
    .add_yaxis(series_name='硬币',y_axis=data['coin'])
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(name="收藏")
    )
)
scatter.render_notebook()
```

```python
graph = (
    Bar(init_opts=opts.InitOpts(width="1080px", height="320px"))
    .add_xaxis(xaxis_data=list(take_20['title'].values))
    .add_yaxis(series_name='播放量',y_axis=list(take_20['view']))
    .add_yaxis(series_name='硬币量',y_axis=list(take_20['coin']))
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name="视频标题",
            axislabel_opts=opts.LabelOpts(rotate=-15)
        ),
        datazoom_opts = opts.DataZoomOpts(is_show=True),
    )
)
graph.render_notebook()
```
