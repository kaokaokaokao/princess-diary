```python
import numpy as np
```

```python
np.zeros((3,4))
# 创建3行4列的空数组
```

```python
np.ones((3,4))
```

```python
# eye函数创建对角线上是1，其他是0的数组
np.eye(3,3)
```

```python
# diag函数创建对角线上为指定数字,其他是0的数组
np.diag([1,3,4,7,8])
```

```python
# 生成随机数 需要调用random子库
# random 生成[0.0，1.0)之间的随机浮点数。参数可以是整数，也可以是元组
print(np.random.random(5))
print(np.random.random((2,3)))
```

```python
# rand 生成[0.0，1.0)之间的均匀分布浮点数组。
print(np.random.rand(2))
print(np.random.rand(2,3))
```

```python
# randint 在指定范围中随机取样，生成随机整数（左闭右开）；uniform 生成随机浮点数
print(np.random.randint(0,100,20))
print(np.random.randint(0,100,size = (3,4)))
print(np.random.uniform(0,100,20))
print(np.random.uniform(0,100,size = (3,4)))
```

```python
# normal 生成正态分布的数组（第一个参数：均值；第二个参数：标准差；第三个参数：生成的维度）
np.random.normal(0,1,(3,4))
```

```python
import matplotlib.pyplot as plt
graph1 = np.random.normal(1.75,1,100000000)
plt.hist(graph1,1000) # hist 直方图
plt.show()# 正态分布图
```

```python
graph2 = np.random.uniform(-1,1,100000000)
plt.hist(graph2,1000) # hist 直方图
plt.show()# 均匀分布图
```

```python
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])
total = arr1 + arr2
total
```

```python
arr1 * 2
```

```python
#比较运算返回 bool值
print(arr2 > arr1)
```

```python
#逻辑运算
print(np.any(arr2 > 4))# 有一个数满足条件即可
print(np.all(arr2 > 7))# 需要所有数都满足条件
```

```python
# mean()求算术平均值
# average()求加权平均值
np.sum(arr1)# 所有数相加
np.max(arr1)# 最大值
np.mean(arr1)# 求算术平均值
```

```python
price = np. array ([34.5,36, 37.8, 39, 39.8, 33.6])
number = np. array ([900, 580, 230, 150, 120, 1800])
print(np.average(price,weights = number))# 加权平均值
print(np.mean(price))# 算术平均值
```

```python
# axis = 0 时，按纵轴计算；axis = 1时，按横轴计算；若为制定，则计算所有
print(np.sum(arr1,axis=0))
print(np.sum(arr1,axis=1))
```

```python
# dot() 矩阵乘法，点乘法，内积运算
np.dot([[1,2,3],[4,5,6]],[2,3,4])
```
