```python
from pyecharts.charts import Scatter,Bar,Line
import pyecharts.options as opts
x = take_20['title'].values
y1 = take_20['view'].values
y2 = take_20['coin'].values
bar = (
    Bar()
    .add_xaxis(list(x))
    .add_yaxis('播放量', list(y1), yaxis_index=0) 
    .extend_axis(yaxis=opts.AxisOpts(name='硬币量'))
)
bar.render_notebook()
```

```python
c = (
    Bar()
    .add_xaxis(list(x))
    .add_yaxis('播放量', list(y1))
    .add_yaxis("硬币量", list(y2))
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            name="视频标题",
            axislabel_opts=opts.LabelOpts(rotate=-15)
        )
    )
)
c.render_notebook()

```
